# Hydrogen Orbitals

The orbitals for a single-electron atom can be solved analytically. However, it is difficult to visualize three-dimensional wavefunctions on paper. In this notebook, we will use Python to make plots of the Hydrogen atom orbitals.

```python
# You don't need to edit anything in this cell.
# Run this cell to define the coordinates and plotting function.

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors

Z = 1      # charge of nucleus (elementary charge units)
a0 = 0.529 # Bohr radius (angstroms)

res = 100 # resolution of image: total voxels = (2*res + 1)^3
x = np.linspace(-10, 10, 2*res+1) # angstroms
y = np.linspace(-10, 10, 2*res+1) # angstroms
z = np.linspace(-10, 10, 2*res+1) # angstroms
xxx, yyy, zzz = np.meshgrid(x, y, z, indexing='ij')
r = np.sqrt(xxx**2 + yyy**2 + zzz**2)

sigma = Z * r / a0
phi = np.arctan2(yyy, xxx)
theta = np.arctan2(np.sqrt(xxx**2 + yyy**2), zzz)

def plot_orbital(Y, R):
    vol = Y(theta, phi) * R(sigma)
    vol[np.abs(vol) < 1e-10] = 0.0 # ignore floating point rounding error
    
    plt.figure(figsize=(12, 3))
    
    plt.subplot(1, 3, 1)
    plt.gca().set_aspect('equal', 'box')
    try:
        plt.pcolormesh(x, y, np.swapaxes(vol[:,:,res],0,1), cmap='RdBu_r', norm=colors.DivergingNorm(0))
    except:
        plt.pcolormesh(x, y, np.swapaxes(vol[:,:,res],0,1), cmap='RdBu_r')
    plt.title(Y.__name__+'*'+R.__name__+' orbital')
    plt.xlabel("x (Å)")
    plt.ylabel("y (Å)")
    plt.colorbar()
    
    plt.subplot(1, 3, 2)
    plt.gca().set_aspect('equal', 'box')
    try:
        plt.pcolormesh(x, z, np.swapaxes(vol[:,res,:],0,1), cmap='RdBu_r', norm=colors.DivergingNorm(0))
    except:
        plt.pcolormesh(x, z, np.swapaxes(vol[:,res,:],0,1), cmap='RdBu_r')
    plt.title(Y.__name__+'*'+R.__name__+' orbital')
    plt.xlabel("x (Å)")
    plt.ylabel("z (Å)")
    plt.colorbar()
    
    plt.subplot(1, 3, 3)
    plt.gca().set_aspect('equal', 'box')
    try:
        plt.pcolormesh(y, z, np.swapaxes(vol[res,:,:],0,1), cmap='RdBu_r', norm=colors.DivergingNorm(0))
    except:
        plt.pcolormesh(y, z, np.swapaxes(vol[res,:,:],0,1), cmap='RdBu_r')
    plt.title(Y.__name__+'*'+R.__name__+' orbital')
    plt.xlabel("y (Å)")
    plt.ylabel("z (Å)")
    plt.colorbar()
        
    plt.tight_layout()
```

```python
# define s orbital wavefunctions

def Ys(theta, phi):
    return (4 * np.pi) ** (-0.5)

def R1s(sigma):
    return 2*(Z / a0) ** (3/2) * np.exp(- Z * sigma / a0)

def R2s(sigma):
    return (1/(2*np.sqrt(2)))*(Z / a0)**(3/2) * (2 - sigma) * np.exp(- sigma / 2)

def R3s(sigma):
    return (1/(81*np.sqrt(3)))*(Z / a0)**(3/2) * (27 - 18*sigma + 2*sigma**2) * np.exp(- sigma / 3)

plot_orbital(Ys, R1s)
plot_orbital(Ys, R2s)
plot_orbital(Ys, R3s)
```
